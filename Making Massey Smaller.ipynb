```python
grouped_counts = Massey.groupby(['SystemName', 'Season']).size().reset_index(name='Counts')
```

```python
pivot_table = grouped_counts.pivot_table(index='SystemName', columns='Season', values='Counts', fill_value=0)

# Plot heatmap
plt.figure(figsize=(30, 36))
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Counts by SystemName and Season')
plt.xlabel('Season')
plt.ylabel('SystemName')
plt.show()
```
